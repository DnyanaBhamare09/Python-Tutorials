```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,2*np.pi,0.01)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="Sin")
plt.plot(x, y2, label="Cos")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

plt.title("Sine and Cosine Function")

plt.show()
```
