```python
import matplotlib.pyplot as plt

activities = ["eat", "sleep", "work", "play"]

slices =[3, 7, 8, 6]

color =["r", "g", "m", "b"]

plt.pie(slices, labels = activities, colors = color)

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

activities = ["eat", "sleep", "work", "play"]

slices =[3, 7, 8, 6]

color =["r", "g", "m", "b"]

plt.pie(slices, labels = activities, colors = color, startangle=90, shadow=True, explode=(0.2,0,0,0))

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

activities = ["eat", "sleep", "work", "play"]

slices =[3, 7, 8, 6]

color =["r", "g", "m", "b"]

plt.pie(slices, labels = activities, colors = color, startangle=90, shadow=True, explode=(0.2,0,0,0), autopct="%1.2f%%" )

plt.legend()

plt.show()
```
