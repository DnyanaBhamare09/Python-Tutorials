```python
import pandas as pd

df = pd.read_csv("employees.csv", parse_dates =["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
```

```python
df.head()
```

```python
mask1 = df["Gender"] == "Male"

mask2 = df["Team"] == "Marketing"

# We want dataframe containing both Male and with Marketing

df[mask1 & mask2].head()
```

```python
mask3 = df["Start Date"] > "1990-01-01"

mask4 = df["Senior Management"] # here we dont need any operative as it is boolen

df[mask3 | mask4].head()  # | operator is taken by "shift+ \ key" This operator gives either of the condition  
```

```python
# we are checking people having name "Robert", work in team of "Client servising" and whoes joining is before 1996
mask5 = df["First Name"] == "Robert"

mask6 = df["Team"] == "Client Services"

mask7 = df["Start Date"] > "1996-01-01"

df[ (mask5 & mask6) | mask7]
```
