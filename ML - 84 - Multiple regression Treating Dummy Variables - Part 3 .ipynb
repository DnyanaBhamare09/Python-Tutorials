```python
import pandas as pd

import matplotlib.pyplot as plt

import numpy as np

from sklearn.preprocessing import LabelEncoder 

from sklearn.preprocessing import OneHotEncoder
```

```python
data = pd.read_csv("Startup_company.csv")

data.head(10)
```

```python
real_x = data.loc[:, "R&D Spend":"State"].values

real_y = data.loc[:, "Profit"].values
```

```python
# Now in state column we have dummy variable. Mainly data is categorised in 3 cities names
# SO first we convert dummy variables into binary variables using transform method i.e LabelEncoder in sklearn
Le = LabelEncoder()
real_x[:,3] = Le.fit_transform(real_x[:,3])

real_x[:, 3]

# Still data is not fully converted into binary. For this we need to pass the data through OneHotEncoder

oneHE = OneHotEncoder(categorical_features= [3])

real_x = oneHE.fit_transform(real_x).toarray()
```
