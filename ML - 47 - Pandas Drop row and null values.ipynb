```python
import pandas as pd

nba = pd.read_csv("nba.csv")
nba
```

```python
nba.dropna()

nba.dropna(how="any")       # removes any NaN value rows. It worls same as dropna()

nba.dropna(how = "all")     # removes rows which has all NaN values
```

```python
# if we want to replace new Dataset from old dataset by removing NaN
nba.dropna(how = "all", inplace= True)
nba
```

```python
nba.dropna(axis = 1)
```

```python
nba.dropna(subset = ["Salary"])
```
