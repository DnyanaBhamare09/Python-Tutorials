```python
import matplotlib.pyplot as plt
import csv
```

```python
x =[]
y =[]

with open("test.csv", "r") as csvfile:
    plots= csv.reader(csvfile)
    for col in plots:
        x.append(col[0])
        y.append(col[1])

plt.plot(x, y, label="File")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.title("Test Graph")

plt.legend()

plt.show()
```
