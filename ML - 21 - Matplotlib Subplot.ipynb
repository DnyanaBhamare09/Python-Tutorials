```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
x1
```

```python
y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.xlabel("x1")
plt.ylabel("Amp(y1)")
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x2, y2, '.-')
plt.title("Subplot")
plt.xlabel("x2")
plt.ylabel("Amp(y2)")
plt.show
```
