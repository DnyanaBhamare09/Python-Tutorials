```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[10,24,36,40,5]

tick_label = ['one','two','three','four','five']
plt.bar(x, y, tick_label = tick_label, width = 0.7)

plt.xlabel("X axis")
plt.ylabel("Y label")
plt.title("Bar Chart")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,24,36,40,5]

tick_label = ['one','two','three','four','five']
plt.bar(x, y, tick_label = tick_label, width = 0.7, color="Green")

plt.xlabel("X axis")
plt.ylabel("Y label")
plt.title("Bar Chart")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,24,36,40,5]

tick_label = ['one','two','three','four','five']
plt.bar(x, y, tick_label = tick_label, width = 0.7, color = ["red", "blue"])

plt.xlabel("X axis")
plt.ylabel("Y label")
plt.title("Bar Chart")
plt.show()
```
